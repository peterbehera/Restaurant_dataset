### 1. Data Exploration and Preprocessing
#### Load the Dataset

```python
# import libraryes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data set
df = pd.read_csv('Dataset .csv')
df.head()
```

### Identify the number of rows and columns

```python
print(f'Number of rows and columns : {df.shape}')

# Total 9551 Rows and 21 Columns
```

### Check for missing values

```python
missing_values = df.isnull().sum()
print(f'Total number of missing values are :')
print(missing_values)
```

### Handle Missing Values

Dropping rows/columns with missing values:

```python
df.dropna(inplace = True)
```

#### Calculate basic statistical measures (mean, median, standard deviation, etc.) for numerical columns.

```python
df.describe()
```

### 2. Analysis of the Target Variable
##### Distribution of "Aggregate rating"

```python
sns.histplot(df['Aggregate rating'], kde=True)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

## Visualize Restaurant Locations on a Map

```python
import folium
from folium.plugins import HeatMap

# Create a base map
base_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

# Add restaurant locations to the map
for idx, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(base_map)

# Save map to an HTML file
base_map.save('restaurant_locations.html')
```

### Visualize Density Using a Heatmap

```python
# Create a heatmap
heat_data = [[row['Latitude'], row['Longitude']] for idx, row in df.iterrows()]

heat_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)
HeatMap(heat_data).add_to(heat_map)

# Save heatmap to an HTML file
heat_map.save('restaurant_heatmap.html')
```

## Analyze Distribution of Restaurants Across Cities or Countries
#### Distribution Across Cities

```python
# Plot the distribution of restaurants across cities
plt.figure(figsize=(12, 6))
sns.countplot(x='City', data=df, order=df['City'].value_counts().index)
plt.title('Distribution of Restaurants Across Cities')
plt.xlabel('city')
plt.xticks(rotation=90)
plt.ylabel('Number of Restaurants')
plt.show()
```

### Distribution Across Countries

```python
# Plot the distribution of restaurants across countries
plt.figure(figsize=(12, 6))
sns.countplot(y='Country Code', data=df, order=df['Country Code'].value_counts().index)
plt.title('Distribution of Restaurants Across Countries')
plt.xlabel('Number of Restaurants')
plt.ylabel('Country Code')
plt.show()
```

## Correlation Between Location and Rating
#### Plotting Rating by Latitude and Longitude

```python
# Scatter plot of latitude and longitude colored by rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', palette='viridis', data=df)
plt.title('Restaurant Locations Colored by Aggregate Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Rating')
plt.show()
```

### Correlation Analysis

```python
# Check correlation between latitude, longitude, and rating
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print(correlation_matrix)
```
